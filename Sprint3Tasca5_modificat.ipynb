```python
# To see which are the Carriers with longer ArrDelay we sort the 50 flights with the longest ArrDelay
```

```python
dfOrg1=dfOrg1.fillna(0)
dfOrg1=dfOrg1.drop(columns=['Unnamed: 0'])
dfOrg1.sort_values(by='ArrDelay',ascending=False).head(50)[["UniqueCarrier","FlightNum","ArrDelay","Origin","Dest"]]
```

```python
# Now we group the flights by carrier so we can see a ranking of the Carriers by the Number Of Flights
```

```python
dfOrg3=dfOrg.rename(columns={"Unnamed: 0": "NumOfFlights"})
dfOrg3=dfOrg3.groupby(['UniqueCarrier']).count()
dfOrg3=dfOrg3[["NumOfFlights"]]
dfOrg3=dfOrg3.reset_index()
dfOrg3
```
